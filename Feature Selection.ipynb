```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.feature_selection import chi2
from sklearn.feature_selection import SelectKBest
```

```python
data=pd.read_csv('/home/girish/Downloads/train.csv')
```

```python
data.head(5)
```

```python
data.shape
```

## Seperating Target Value And Features

```python
X=data.iloc[:,:20]
y=data.iloc[:,-1]
```

```python
X.head(5)
```

```python
y.head(5)
```

```python
y=pd.DataFrame(y)
```

```python
type(y)
```

```python
y.head(2)
```

# Feature Selection using chi2

```python
best_features = SelectKBest(score_func=chi2,k=10)
```

```python
fit = best_features.fit(X,y)
```

```python
fit.scores_
```

```python
fitData=pd.DataFrame(fit.scores_)
```

```python
fitData.columns = ['Score']
fitData.sort_values('Score',ascending=False)
```

```python
X.columns
```

```python
Xcolumns=pd.DataFrame(X.columns)
```

```python
Xcolumns
```

```python
featureScore=pd.concat([Xcolumns,fitData],axis=1)
```

```python
featureScore.columns=['features','Score']
```

```python
ten_features = featureScore.sort_values(by='Score',ascending=False).head(10)['features'].values
```

```python
ten_features
```

# Feature Importance
### Tree based Method


```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier()
```

```python
model.fit(X,y)
```

```python
model.feature_importances_ 
```

```python
feature_Importance_Data=pd.DataFrame(model.feature_importances_ ,index=X.columns,columns=['Importance'])
```

```python
feature_Importance_Data=feature_Importance_Data.sort_values(by='Importance',ascending=False)
```

```python
plt.figure(figsize=(20,10))
plt.bar(feature_Importance_Data.index,feature_Importance_Data['Importance'])
plt.show()
```
