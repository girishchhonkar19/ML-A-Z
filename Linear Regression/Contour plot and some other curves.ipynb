```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a=np.arange(-1,1,0.02)
b=a
```

```python
a,b=np.meshgrid(a,b)
```

```python
print(a)
```

```python
print(b)
```

```python
fig=plt.figure()
axes = fig.gca(projection='3d')
axes.plot_surface(a,b,a*a+b*b,cmap='rainbow')
plt.show()
```

```python
fig=plt.figure()

axes = fig.gca(projection='3d')

axes.contour(a,b,a**2 + b**2,cmap='rainbow')
plt.show()
```
